```python
#cleaning the wrong format of data
df = pd.read_csv("dirtydata.csv")
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = 'mixed')
print(df)
```

```python
df.dropna(subset=['Date','Calories'], inplace = True)
print(df)
```

```python
df.loc[7,'Duration'] = 150
print(df)
```

```python
for x in df.index:
  if df.loc[x,"Duration"] > 70:
    df.loc[x, "Duration"] = 70
print(df)
```

```python
for x in df.index:
  if df.loc[x, 'Duration'] > 60:
    df.drop(x , inplace = True)
print(df)
```
